# Relative distribution of item weight and its visibility(scatter plot)


```python
import seaborn as sns

sns.scatterplot(x=df['Item_Weight'], y=df['Item_Visibility'])
plt.title("Scatter Plot of Item Weight vs Item Visibility")
plt.show()
```


# Violin plot for Item_Outlet_Sales


```python
sns.violinplot(df['Item_Outlet_Sales'])
```

# sctter plot using seaborn add hue

```python
df5 = df[['Item_Weight','Item_MRP','Item_Outlet_Sales','Outlet_Type']]

sns.violinplot(x='Outlet_Type', y='Item_Outlet_Sales', hue='Outlet_Type', data=df5, split=False)
```

# heat map for selected numeric columns

```python
num_cols = ['Item_Weight','Item_MRP','Item_Outlet_Sales']

corr = df[num_cols].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
```
